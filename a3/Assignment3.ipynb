This is the image that the algorithm performs well

The parameter RandomPatchSD is the standard deviation of the number of the patch that gets chosen. It determines how many patches are included in the random selection of the matched texture regions. If the RandomPatchSD gets too small, which means 0, then the optimal patch is always chosen. If the RandomPatchSD gets too large, then too many patches will be included in the selection which might result that some patches of texture with large SSD would be chosen. As a consequence, the algorithm will lead to a poor synthesized image.

The parameter patchL is a half of the length of the synthesized patch - 1. Therefore, it determines the size of the synthesized patchs. If PatchL gets too small, the patch size might be too small that the key structure and the pattern of the texture are missed entirely by the patch. If the PatchL gets too large, it might capture too much information about a particular patch that should not appear to frequently, which might lead to repetition instead of texture synthesis.
